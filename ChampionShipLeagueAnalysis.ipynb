### 3.2 Ev Sahibi ve Deplasman TakÄ±mlarÄ± ArasÄ±ndaki Fark Ä°nceleme

###### Bu adÄ±mda, ev sahibi ve deplasman takÄ±mlarÄ±nÄ±n performanslarÄ± arasÄ±nda anlamlÄ± bir fark olup olmadÄ±ÄŸÄ±nÄ± test edebiliriz.
###### Bu tÃ¼r bir analiz iÃ§in BaÄŸÄ±msÄ±z Ä°ki Ã–rneklem t-testi (Independent Two-Sample t-test) kullanabiliriz. 
###### Bu test, iki grup arasÄ±ndaki ortalama farkÄ±nÄ±n istatistiksel olarak anlamlÄ± olup olmadÄ±ÄŸÄ±nÄ± belirlememize yardÄ±mcÄ± olur.

##### Burada, ev sahibi takÄ±mlarÄ±nÄ±n attÄ±ÄŸÄ± gollerle deplasman takÄ±mlarÄ±nÄ±n attÄ±ÄŸÄ± goller arasÄ±ndaki farkÄ± test edeceÄŸiz. Hedefimiz, her iki takÄ±m arasÄ±ndaki gol ortalamalarÄ±nÄ±n birbirinden anlamlÄ± bir ÅŸekilde farklÄ± olup olmadÄ±ÄŸÄ±nÄ± gÃ¶rmek.

```python
from scipy import stats

#Ev sahibi ve deplasman golleri karÅŸÄ±laÅŸtÄ±rma
home_goals = df["FTH Goals"]
away_goals = df["FTA Goals"]

#BaÄŸÄ±msÄ±z Ä°ki Ã¶rneklem t-testi
t_start,p_value = stats.ttest_ind(home_goals,away_goals)

#SonuÃ§larÄ± yazdÄ±rma
print("t-deÄŸeri",t_start)
print("p-deÄŸeri",p_value)
```

### 3.2.1 Ev Sahibi ve Deplasman PerformansÄ± ArasÄ±ndaki Fark
##### Bu adÄ±mda, ev sahibi takÄ±mlarÄ±nÄ±n ve deplasman takÄ±mlarÄ±nÄ±n attÄ±ÄŸÄ± goller arasÄ±ndaki farkÄ± inceledik. Ä°ki grup arasÄ±ndaki farkÄ±n anlamlÄ± olup olmadÄ±ÄŸÄ±nÄ± deÄŸerlendirmek iÃ§in BaÄŸÄ±msÄ±z Ä°ki Ã–rneklem t-testi kullandÄ±k.

### 3.2.2 t-Testi SonuÃ§larÄ±
##### Ev sahibi ve deplasman gollerinin istatistiksel olarak karÅŸÄ±laÅŸtÄ±rÄ±lmasÄ± sonucunda elde edilen t-deÄŸeri ve p-deÄŸerine gÃ¶re, gruplar arasÄ±ndaki farkÄ±n anlamlÄ± olup olmadÄ±ÄŸÄ±na karar verdik.

##### t_deÄŸeri =20.64 Ev sahÅŸbÅŸ takÄ±mlarÄ±nÄ±n attÄ±ÄŸÄ± goller ile deplasman takÄ±mlarÄ±nÄ±n attÄ±ÄŸÄ± goller arasÄ±ndaki farkÄ±n oldukÃ§a bÃ¼yÃ¼k olduÄŸunu gÃ¶steriyor

##### p_deÄŸeri =0.05 'ten Ã§ok daha kÃ¼Ã§Ã¼k bir deÄŸer olduÄŸu iÃ§in ,ev sahibi ve deplasman takÄ±mlarÄ±nÄ±n ortlama gol sayÄ±larÄ± arasÄ±nda istatiksel olarak anlamlÄ± bir fark olduÄŸunu sÃ¶yleyebiliriz. Ev sahibi takÄ±mlar genellikle daha fazla gol atar

### 6. Åžut etkinliÄŸi analizi: En fazla ÅŸut Ã§eken ama az gol atan takÄ±mlar

```python
# Åžut etkinliÄŸini ev sahibi ve deplasman takÄ±mlarÄ± iÃ§in hesaplayalÄ±m
df['Home Shot Efficiency'] = df['FTH Goals'] / df['H Shots']
df['Away Shot Efficiency'] = df['FTA Goals'] / df['A Shots']

# Ev sahibi ve deplasman takÄ±mlarÄ±nÄ±n ÅŸut etkinliÄŸini karÅŸÄ±laÅŸtÄ±ralÄ±m
df['Shot Efficiency'] = (df['Home Shot Efficiency'] + df['Away Shot Efficiency']) / 2

# Null ya da sonsuz deÄŸere sahip satÄ±rlarÄ± filtreleyelim
df_clean = df[df['Shot Efficiency'].notnull() & (df['Shot Efficiency'] != float('inf'))]

# Åžut etkinliÄŸi verilerini tekrar kontrol edelim
print(df_clean[['HomeTeam', 'AwayTeam', 'Shot Efficiency']].head())

# Åžut etkinliÄŸi en dÃ¼ÅŸÃ¼k olan takÄ±mlarÄ± bulalÄ±m (az gol atan ama Ã§ok ÅŸut Ã§eken takÄ±mlar)
least_efficiency_shooters = df_clean[['HomeTeam', 'AwayTeam', 'Shot Efficiency']].sort_values(by='Shot Efficiency').head(10)

# GÃ¶rselleÅŸtirme
plt.figure(figsize=(10, 6))
least_efficiency_shooters['Shot Efficiency'].plot(kind='barh', color='red')
plt.title('En Fazla Åžut Ã‡eken Ama Az Gol Atan TakÄ±mlar')
plt.xlabel('Åžut EtkinliÄŸi (Gol / Åžut SayÄ±sÄ±)')
plt.ylabel('TakÄ±mlar')
plt.show()

```

```python
df["Total Fouls"] = df["H Fouls"]+df["A Fouls"]

# Hakemlerin yÃ¶nettiÄŸi maÃ§ baÅŸÄ±na toplam faul sayÄ±sÄ±nÄ± hesaplama

referee_fouls = df.groupby("Referee")["Total Fouls"].sum()

# Hakemlerin yÃ¶nettiÄŸi maÃ§ sayÄ±sÄ±nÄ± hesaplama
referee_matches = df.groupby("Referee").size()

# Hakemlerin yÃ¶nettiÄŸi maÃ§ baÅŸÄ±na ortalama faul sayÄ±sÄ±nÄ± hesaplama
# Hakemlerin yÃ¶nettiÄŸi maÃ§ baÅŸÄ±na ortalama faul sayÄ±sÄ±nÄ± hesaplama
referee_avg_fouls = referee_fouls / referee_matches

# Ortalama faul sayÄ±sÄ±nÄ± bÃ¼yÃ¼kten kÃ¼Ã§Ã¼ÄŸe sÄ±ralama
referee_avg_fouls = referee_avg_fouls.sort_values(ascending=False)

referee_avg_fouls
```

```python
# Kruskal-Wallis testi
kruskal_yellow = stats.kruskal(*[referee_cards['Total Yellow Cards'][referee_cards['Total Yellow Cards'] == referee] for referee in referee_cards.index])
kruskal_red = stats.kruskal(*[referee_cards['Total Red Cards'][referee_cards['Total Red Cards'] == referee] for referee in referee_cards.index])
kruskal_fouls = stats.kruskal(*[referee_fouls[referee_fouls == referee] for referee in referee_fouls.index])

# SonuÃ§larÄ± yazdÄ±ralÄ±m
print("SarÄ± Kart Kruskal-Wallis Sonucu:", kruskal_yellow)
print("KÄ±rmÄ±zÄ± Kart Kruskal-Wallis Sonucu:", kruskal_red)
print("Faul Kruskal-Wallis Sonucu:", kruskal_fouls)
```
